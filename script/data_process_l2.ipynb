# `orderbook_preprocess(df)`

Converts a raw Databento MBP-10 event log into two outputs: a snapshot dictionary of the order book state at each timestamp, and a delta list describing what changed between consecutive snapshots.

---

## Inputs

| Parameter | Type | Description |
|---|---|---|
| `df` | `pd.DataFrame` | Raw event log grouped by `ts_recv`. Must contain Databento MBP-10 fields: `flags`, `action`, `side`, `bid_px_00..03`, `bid_sz_00..03`, `ask_px_00..03`, `ask_sz_00..03` |

---

## Outputs

| Return value | Type | Description |
|---|---|---|
| `book_states` | `dict[ts -> dict]` | Full order book snapshot at each valid timestamp. Keys are `bid_03..bid_00, ask_00..ask_03`, each storing `[price, size]` |
| `delta` | `list[ts, level, size, action]` | One entry per detected change. `level` follows the paper's relative indexing: `-4..−1` for bid, `+1..+4` for ask |

---

## Processing steps

**1. Group by timestamp** — events sharing the same `ts_recv` are processed together as one atomic book update.

**2. Record book state** — for each group whose last row has `flags ∈ {128, 130}` and `action ≠ "T"`, the top-4 bid and ask levels are snapshotted into `book_states[ts]`.

**3. Classify the event type** — two paths depending on whether `"N"` (None/no side) appears in the group's side column:

- **`"N"` present → non-trade event** (Add, Cancel, Modify). The current snapshot is compared against the previous one (`pre_ts`) to compute the delta. Two sub-cases:
  - *Prices unchanged* — only sizes shifted, recorded as `"A"` (increase) or `"C"` (decrease) per level.
  - *Prices changed* — `scan_side()` scans bid and ask separately from best outward. The first price difference found determines the event type (Create, price move, or gap fill) and all deeper levels are interpreted relative to that.

- **`"N"` absent → trade event**. Each row in the group is recorded directly into `delta` using its `depth` field to determine the queue level. A trailing Cancel on normal trades (`flags 128/130`, last action `"C"` preceded by `"T"`) is stripped before recording.

**4. Advance `pre_ts`** — after processing, `pre_ts` is updated to `ts` so the next iteration can compare against the current snapshot.

---

## Queue level convention

```
i=0 → q−4   (deepest bid)
i=1 → q−3
i=2 → q−2
i=3 → q−1   (best bid)
i=4 → q+1   (best ask)
i=5 → q+2
i=6 → q+3
i=7 → q+4   (deepest ask)
```

Negative levels are bid-side, positive levels are ask-side. This matches the relative indexing used throughout the simulator — levels shift after every price move rather than being anchored to an absolute price.

---

## Known limitations

- `row.depth` in the trade branch assumes this field exists on the input DataFrame. Verify it is present or replace with a price-to-level lookup against the current snapshot.
- Only timestamps with `flags ∈ {128, 130}` are recorded into `book_states`. Timestamps that fail this filter are skipped entirely and do not advance `pre_ts`, which means the delta comparison always uses the last valid snapshot as the baseline.
- The function processes one ticker at a time. For multi-ticker DataFrames, filter by instrument ID before calling.

# `convert_book_states`

Converts the raw `book_states` dictionary produced by the preprocessing loop into a structured DataFrame ready for parameter estimation.

---

## Input

`book_states` is a dict keyed by `ts_recv` (nanosecond UTC integer), where each value is a snapshot of the top 4 bid and ask levels at that timestamp:

```python
{
    1704891000000000000: {
        "bid_00": [price, size],   # best bid
        "bid_01": [price, size],   # second best bid
        "bid_02": [price, size],
        "bid_03": [price, size],
        "ask_00": [price, size],   # best ask
        "ask_01": [price, size],
        "ask_02": [price, size],
        "ask_03": [price, size],
    },
    ...
}
```

Prices are in Databento fixed-point units (int64, 1e-10 dollars). One cent = `1_000_000_000` units, which is the default `tick_size`.

---

## What it does

**Step 1 — Compute MES per depth level.** The Median Event Size (MES) is calculated separately for each of the 4 depth slots by pooling bid and ask sizes at the same depth across all timestamps. Bid and ask are symmetrised at each level — consistent with the paper's estimation procedure. MES is used to normalise raw share counts into dimensionless queue units so that a queue of 1.0 always represents one median-sized order regardless of depth.

**Step 2 — Compute spread and imbalance.** For each snapshot, spread is computed as `(best_ask - best_bid) / tick_size` in ticks. Imbalance is `(bid_sz_00 - ask_sz_00) / (bid_sz_00 + ask_sz_00)`, ranging from −1 (fully ask-heavy) to +1 (fully bid-heavy). Snapshots where either best price is missing or NaN are skipped.

**Step 3 — Re-index queues by tick distance.** Rather than using the raw slot indices (`bid_00`, `bid_01`, ...), queues are re-indexed by their price distance from the current best. `q-1` is always the best bid regardless of its absolute price; `q-2` is exactly one tick behind the best bid; and so on. This produces the relative indexing scheme the simulator depends on — where `q±k` shifts with the market rather than being anchored to a fixed reference price.

**Step 4 — Normalise volumes.** Each queue size is divided by the MES for its depth level, producing normalised units. A value of `2.0` at `q-1` means the best bid queue holds twice the median order size.

---

## Output

A DataFrame with one row per timestamp and the following columns:

| Column | Type | Description |
|---|---|---|
| `ts` | int64 | Exchange timestamp in nanoseconds UTC |
| `spread` | int | Bid-ask spread in ticks |
| `imbalance` | float | Best-level imbalance, range [−1, 1] |
| `best_size` | float | Combined best bid + ask size, normalised |
| `q-4` | float | Fourth-best bid queue, MES-normalised |
| `q-3` | float | Third-best bid queue, MES-normalised |
| `q-2` | float | Second-best bid queue, MES-normalised |
| `q-1` | float | Best bid queue, MES-normalised |
| `q+1` | float | Best ask queue, MES-normalised |
| `q+2` | float | Second-best ask queue, MES-normalised |
| `q+3` | float | Third-best ask queue, MES-normalised |
| `q+4` | float | Fourth-best ask queue, MES-normalised |

---

## Notes

- Snapshots with a missing or NaN best bid or ask are silently dropped. Monitor how many rows are skipped — a high drop rate indicates data quality issues in the preprocessing step upstream.
- `best_size` is scaled by `1e-9` to convert from Databento fixed-point units to a human-readable share count. The other queue columns are in MES units and do not need this conversion.
- The `columns=` argument passed to `pd.DataFrame()` at the end does not include `best_size` — it will be silently dropped from the result. Add it to the list if you need it downstream.

# Event Aggregation

## `aggregate_trades`

Groups consecutive Trade rows sharing the same `ts` into one row with summed size.

**Why**: a single aggressive order hitting multiple resting orders at the same price
generates one fill message per resting order. These must be treated as one event for
volume estimation — otherwise the volume distribution is artificially fragmented.

**Trigger**: `action == "T"` rows with identical `ts`.

**Key column**: `size` is summed; all other columns take the first row's value.

```
Before                          After
ts=100  T  +1  200             ts=100  T  +1  600   ← 200+250+150
ts=100  T  +1  250    ──────►
ts=100  T  +1  150
ts=101  T  +1  300             ts=101  T  +1  300   ← untouched
```

---

## `aggregate_creates`

Folds Add events at the same `ts` and `level` into their preceding CREATE event.

**Why**: when a new price level is created, multiple participants immediately join it
in the same timestamp. Estimating Create volume from the first message alone
underestimates the true initial queue size.

**Trigger**: `action in ["CREATE_ASK", "CREATE_BID"]` rows sharing `ts` and `level`.

**Key column**: `size` is summed across all participants joining the new level.

```
Before                              After
ts=200  CREATE_ASK  +1  100        ts=200  CREATE_ASK  +1  350  ← 100+150+100
ts=200  CREATE_ASK  +1  150  ────►
ts=200  CREATE_ASK  +1  100
ts=201  A           +2   50        ts=201  A           +2   50  ← untouched
```



````python
def aggregation_stats(raw_df: pd.DataFrame, 
                      agg_trades_df: pd.DataFrame,
                      agg_creates_df: pd.DataFrame) -> None:
    """
    Print before/after statistics for both aggregation steps.
    raw_df         : delta_df before any aggregation
    agg_trades_df  : output of aggregate_trades()
    agg_creates_df : output of aggregate_creates()
    """
    raw_trades  = raw_df[raw_df["action"] == "T"]
    raw_creates = raw_df[raw_df["action"].isin(["CREATE_ASK", "CREATE_BID"])]
    agg_trades  = agg_trades_df[agg_trades_df["action"] == "T"]
    agg_creates = agg_creates_df[agg_creates_df["action"].isin(["CREATE_ASK", "CREATE_BID"])]

    print("=" * 50)
    print("TRADE AGGREGATION")
    print("=" * 50)
    print(f"  Raw trade rows      : {len(raw_trades):>8,}")
    print(f"  Aggregated rows     : {len(agg_trades):>8,}")
    print(f"  Rows removed        : {len(raw_trades) - len(agg_trades):>8,}")
    print(f"  Compression ratio   : {len(raw_trades)/len(agg_trades):.2f}x")
    print(f"  Total volume before : {raw_trades['size'].sum():>8,}")
    print(f"  Total volume after  : {agg_trades['size'].sum():>8,}")
    print(f"  Volume preserved    : {raw_trades['size'].sum() == agg_trades['size'].sum()}")
    print(f"  Avg burst size      : {len(raw_trades)/len(agg_trades):.2f} fills per event")
    # Size distribution of aggregated trades
    print(f"\n  Aggregated trade size distribution:")
    print(agg_trades["size"].describe().to_string())

    print()
    print("=" * 50)
    print("CREATE AGGREGATION")
    print("=" * 50)
    print(f"  Raw create rows     : {len(raw_creates):>8,}")
    print(f"  Aggregated rows     : {len(agg_creates):>8,}")
    print(f"  Rows removed        : {len(raw_creates) - len(agg_creates):>8,}")
    print(f"  Compression ratio   : {len(raw_creates)/max(len(agg_creates),1):.2f}x")
    print(f"  Total volume before : {raw_creates['size'].sum():>8,}")
    print(f"  Total volume after  : {agg_creates['size'].sum():>8,}")
    print(f"  Volume preserved    : {raw_creates['size'].sum() == agg_creates['size'].sum()}")

    # Split by side
    for action in ["CREATE_ASK", "CREATE_BID"]:
        subset = agg_creates[agg_creates["action"] == action]
        print(f"\n  {action} size distribution ({len(subset):,} events):")
        print(subset["size"].describe().to_string())

    print()
    print("=" * 50)
    print("OVERALL EVENT MIX (after both aggregations)")
    print("=" * 50)
    counts = agg_creates_df["action"].value_counts()
    total  = len(agg_creates_df)
    for action, count in counts.items():
        print(f"  {action:<15} {count:>8,}  ({100*count/total:.2f}%)")
````

````python
from IPython.display import display, HTML
import pandas as pd

def aggregation_stats_fulldisplay(raw_df: pd.DataFrame,
                      agg_trades_df: pd.DataFrame,
                      agg_creates_df: pd.DataFrame) -> None:

    raw_trades  = raw_df[raw_df["action"] == "T"]
    raw_creates = raw_df[raw_df["action"].isin(["CREATE_ASK", "CREATE_BID"])]
    agg_trades  = agg_trades_df[agg_trades_df["action"] == "T"]
    agg_creates = agg_creates_df[agg_creates_df["action"].isin(["CREATE_ASK", "CREATE_BID"])]

    vol_ok_t = raw_trades["size"].sum() == agg_trades["size"].sum()
    vol_ok_c = raw_creates["size"].sum() == agg_creates["size"].sum()

    trade_desc  = agg_trades["size"].describe()
    ask_desc    = agg_creates[agg_creates["action"] == "CREATE_ASK"]["size"].describe()
    bid_desc    = agg_creates[agg_creates["action"] == "CREATE_BID"]["size"].describe()

    counts = agg_creates_df["action"].value_counts()
    total  = len(agg_creates_df)
    mix_rows = "".join(
        f"<tr><td style='padding:3px 8px;color:var(--color-text-secondary)'>{a}</td>"
        f"<td style='padding:3px 8px;text-align:right'>{c:,}</td>"
        f"<td style='padding:3px 8px;text-align:right;color:var(--color-text-secondary)'>{100*c/total:.1f}%</td></tr>"
        for a, c in counts.items()
    )

    def desc_rows(s):
        keep = ["mean", "std", "min", "50%", "max"]
        return "".join(
            f"<tr><td style='padding:2px 8px;color:var(--color-text-secondary);font-size:12px'>{k}</td>"
            f"<td style='padding:2px 8px;text-align:right;font-size:12px'>{s[k]:,.1f}</td></tr>"
            for k in keep if k in s
        )

    def metric(label, value, sub=None, ok=None):
        badge = ""
        if ok is not None:
            col = "var(--color-background-success)" if ok else "var(--color-background-danger)"
            tc  = "var(--color-text-success)" if ok else "var(--color-text-danger)"
            badge = f"<span style='font-size:11px;padding:2px 6px;border-radius:4px;background:{col};color:{tc};margin-left:6px'>{'✓' if ok else '✗'}</span>"
        sub_html = f"<div style='font-size:12px;color:var(--color-text-secondary);margin-top:2px'>{sub}</div>" if sub else ""
        return (
            f"<div style='background:var(--color-background-secondary);border-radius:8px;"
            f"padding:12px 16px;min-width:120px'>"
            f"<div style='font-size:12px;color:var(--color-text-secondary);margin-bottom:4px'>{label}</div>"
            f"<div style='font-size:20px;font-weight:500'>{value}{badge}</div>"
            f"{sub_html}</div>"
        )

    html = f"""
    <div style='font-family:var(--font-sans);padding:4px 0'>

      <div style='font-size:13px;font-weight:500;margin:0 0 8px;color:var(--color-text-secondary);
                  text-transform:uppercase;letter-spacing:.04em'>Trade aggregation</div>
      <div style='display:flex;flex-wrap:wrap;gap:10px;margin-bottom:16px'>
        {metric("Raw rows",       f"{len(raw_trades):,}")}
        {metric("Agg rows",       f"{len(agg_trades):,}")}
        {metric("Rows removed",   f"{len(raw_trades)-len(agg_trades):,}")}
        {metric("Compression",    f"{len(raw_trades)/max(len(agg_trades),1):.2f}×")}
        {metric("Volume",         f"{agg_trades['size'].sum():,}", "after agg", ok=vol_ok_t)}
        {metric("Avg burst",      f"{len(raw_trades)/max(len(agg_trades),1):.2f}", "fills/event")}
      </div>
      <details style='margin-bottom:20px'>
        <summary style='font-size:12px;color:var(--color-text-secondary);cursor:pointer'>
          Trade size distribution
        </summary>
        <table style='font-size:13px;margin-top:6px;border-collapse:collapse'>
          {desc_rows(trade_desc)}
        </table>
      </details>

      <div style='font-size:13px;font-weight:500;margin:0 0 8px;color:var(--color-text-secondary);
                  text-transform:uppercase;letter-spacing:.04em'>Create aggregation</div>
      <div style='display:flex;flex-wrap:wrap;gap:10px;margin-bottom:16px'>
        {metric("Raw rows",     f"{len(raw_creates):,}")}
        {metric("Agg rows",     f"{len(agg_creates):,}")}
        {metric("Rows removed", f"{len(raw_creates)-len(agg_creates):,}")}
        {metric("Compression",  f"{len(raw_creates)/max(len(agg_creates),1):.2f}×")}
        {metric("Volume",       f"{agg_creates['size'].sum():,}", "after agg", ok=vol_ok_c)}
      </div>
      <details style='margin-bottom:20px'>
        <summary style='font-size:12px;color:var(--color-text-secondary);cursor:pointer'>
          Size by side
        </summary>
        <div style='display:flex;gap:24px;margin-top:8px'>
          <div>
            <div style='font-size:12px;font-weight:500;margin-bottom:4px'>CREATE_ASK
              ({len(agg_creates[agg_creates["action"]=="CREATE_ASK"]):,})</div>
            <table style='font-size:12px;border-collapse:collapse'>{desc_rows(ask_desc)}</table>
          </div>
          <div>
            <div style='font-size:12px;font-weight:500;margin-bottom:4px'>CREATE_BID
              ({len(agg_creates[agg_creates["action"]=="CREATE_BID"]):,})</div>
            <table style='font-size:12px;border-collapse:collapse'>{desc_rows(bid_desc)}</table>
          </div>
        </div>
      </details>

      <div style='font-size:13px;font-weight:500;margin:0 0 8px;color:var(--color-text-secondary);
                  text-transform:uppercase;letter-spacing:.04em'>Event mix (after aggregation)</div>
      <table style='font-size:13px;border-collapse:collapse'>
        <thead>
          <tr>
            <th style='padding:3px 8px;text-align:left;font-weight:500;
                       color:var(--color-text-secondary)'>action</th>
            <th style='padding:3px 8px;text-align:right;font-weight:500;
                       color:var(--color-text-secondary)'>count</th>
            <th style='padding:3px 8px;text-align:right;font-weight:500;
                       color:var(--color-text-secondary)'>share</th>
          </tr>
        </thead>
        <tbody>{mix_rows}</tbody>
      </table>
    </div>
    """
    display(HTML(html))
````
